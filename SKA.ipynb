```python
import matplotlib.pyplot as plt
import numpy as np

w = 5e-8

x = np.linspace(-w,w,1000)
y = (1 - np.cos(x))/x**2


# Insert a NaN where the difference between successive points is negative
y[:-1][(np.diff(y) == 0) | (np.abs(np.diff(y)) >0.9) ] = np.nan

plt.xlim(-w,w)
plt.ylim(-0.1,1.1)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Choose evenly spaced x intervals
x = np.arange(-2*np.pi, 2*np.pi, 0.1)

# plot y = tan(x)
#plt.plot(x, np.tan(x))

# Compute the tangent for each point
y = np.tan(x)

# Insert a NaN where the difference between successive points is negative
y[:-1][np.diff(y) < 0] = np.nan

# Plot the resulting discontinuous line
plt.plot(x, y)
# Set the range of the axes
plt.axis([-2*np.pi, 2*np.pi, -2, 2])

# Include a title
plt.title('y = tan(x)')

# Optional grid-lines
plt.grid()

# Show the graph
plt.show()
```
