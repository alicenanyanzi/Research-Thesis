```python
import networkx as nx

G = nx.Graph()
e_list =[(1,2),(1,4),(2,3),(1,3),(3,4),(3,5),(4,5)]
G.add_edges_from(e_list)
nx.edge_betweenness_centrality(G,normalized=False, weight=None)
```

```python
import networkx as nx

G = nx.DiGraph()
e_list =[(1,2,90),(1,3,110),(3,2,10),(2,4,60),(2,5,30),(3,5,30),(4,5,40),(4,6,50),(5,6,100)]
G.add_weighted_edges_from(e_list)
nx.edge_betweenness_centrality(G,normalized=False, weight='weight')
```
