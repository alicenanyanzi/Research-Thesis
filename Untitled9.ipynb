```python
import networkx as nx
import numpy as np
from scipy.linalg import expm
import matplotlib.pyplot as plt
from pylab import *
import random

G = nx.Graph()
e_list = [(2,1),(3,1),(1,4),(1,5)]
G.add_edges_from(e_list)
a = nx.laplacian_matrix(G).todense()

phi_0 = [3,2,0,0,0]
c_list = []

for t in arange(0,1,0.05):
    phi_t = np.dot(phi_0, expm(-a*t*1))
    c_list.append(phi_t)

plt.plot(arange(0,1,0.05),c_list,marker='*') 
plt.title("Difussion over a network")
plt.xlabel("Time")
plt.ylabel("Average quantity for heat")
plt.show()
```
